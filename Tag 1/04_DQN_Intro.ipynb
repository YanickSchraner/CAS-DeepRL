DQN
---
Train a Flatland-RL policy using DQN

```
import os
from argparse import Namespace
from pprint import pprint

from train import train_agent
from utils.dddqn_policy import DDDQNPolicy
```

Set policy class to the DDDQN policy:

```
policy_class = DDDQNPolicy
```

Training parameters, and environment parameters.

We'll go through them together to see what and why to change each of them!

```
training_params = {
    'n_episodes': 5000,
    'n_evaluation_episodes': 5,
    'checkpoint_interval': 150,
    'eps_start': 1.0,
    'eps_end': 0.01,
    'eps_decay': 0.95,
    'buffer_size': int(1e5),
    'buffer_min_size': 0,
    'batch_size': 128,
    'gamma': 0.95,
    'tau': 1e-3,
    'learning_rate': 1e-4,
    'hidden_layers': [32, 32],
    'update_every': 4,
    'use_gpu': True,
    'num_threads': 1,
    'render': False, # won't work in notebook
    'load_policy': '' # 'runs/2021-09-29_12-47-37/ckpts/ckpt-300.pth',
}
```

```
training_env_params = {
    "n_agents": 5,
    "x_dim": 30,
    "y_dim": 30,
    "n_cities": 2,
    "max_rails_between_cities": 2,
    "max_rail_pairs_in_city": 3,
    "malfunction_rate": 1 / 50,
    "seed": 0
}
evaluation_env_params = training_env_params
```

Observation Parameters

```
# CAUTION - don't change these parameters!
# Your agents will be evaluated with the values below, so you shouldn't change them during training.
obs_params = {
    "observation_tree_depth": 2,
    "observation_radius": 10,
    "observation_max_path_depth": 30
}
```

Run Training

```
from train import train_agent
```

```
print("\nTraining parameters:")
pprint(training_params)

print("\nTraining environment parameters:")
pprint(training_env_params)

print("\nEvaluation environment parameters:")
pprint(evaluation_env_params)

print("\nObservation parameters:")
pprint(obs_params)

os.environ["OMP_NUM_THREADS"] = str(training_params['num_threads'])

train_agent(policy_class,
            Namespace(**training_params),
            Namespace(**training_env_params),
            Namespace(**evaluation_env_params),
            Namespace(**obs_params))
```
